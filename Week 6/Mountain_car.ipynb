```python
import gym
import numpy as np
from matplotlib import pyplot as plt
```

```python
env = gym.make("MountainCar-v0")
```

```python
env.reset()
plt.figure()
plt.imshow(env.render(mode='rgb_array'))

[env.step(0) for x in range(10000)]
plt.figure()
plt.imshow(env.render(mode='rgb_array'))

env.close()
```
